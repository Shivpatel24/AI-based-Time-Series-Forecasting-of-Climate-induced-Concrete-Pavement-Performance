```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Load CSV
file_path = '/home/bel/Desktop/Shiv_SRIP/ATenLoc/SARIMA/output_sarima.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Extract Actual and Predicted
y_true = df['Actual_TL']
y_pred = df['Predicted_TL']

# Calculate metrics
r2 = r2_score(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f'RÂ²: {r2:.4f}')
print(f'MAE: {mae:.4f}')
print(f'RMSE: {rmse:.4f}')

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, color='blue', alpha=0.6, label='Predictions')
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Numerical Model TL')
plt.ylabel('ML model TL')
plt.title('TL')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# === Load your CSV file ===
csv_path = '/home/bel/Desktop/Shiv_SRIP/ATenLoc/SARIMA/output_sarimacesr.csv'  # ð Update with your actual path
df = pd.read_csv(csv_path)

# === Clean the data ===
# Replace infinities with NaN and drop any invalid rows
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['Actual_CESR', 'Predicted_CESR'])

# Keep only rows where both Actual and Predicted CESR are in [-1, 1]
df = df[(df['Actual_CESR'].between(-1, 1)) & (df['Predicted_CESR'].between(-1, 1))]

# Round predicted values to 4 decimals
df['Predicted_CESR'] = df['Predicted_CESR'].round(4)

# === Extract values ===
y_true = df['Actual_CESR'].values
y_pred = df['Predicted_CESR'].values

# === Debug: Check value ranges ===
print("Max Actual_CESR:", np.max(np.abs(y_true)))
print("Max Predicted_CESR:", np.max(np.abs(y_pred)))

# === Compute metrics ===
r2 = r2_score(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print("\n--- Evaluation Metrics ---")
print(f"RÂ² Score: {r2:.4f}")
print(f"MAE:      {mae:.4f}")
print(f"RMSE:     {rmse:.4f}")

# === Scatter Plot ===
plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, color='blue', alpha=0.6, label='Predictions')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], 'r--', label='Ideal Line')
plt.xlabel('Numerical Model CESR')
plt.ylabel(' SARIMAX model CESR ')
plt.title('CESR')
plt.grid(True)
plt.legend()
try:
    plt.tight_layout()
except Exception as e:
    print("tight_layout() failed:", e)
plt.show()
```
